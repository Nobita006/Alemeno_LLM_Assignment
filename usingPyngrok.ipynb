```python
import time
from pyngrok import ngrok
import threading
import os

def run_streamlit():
    os.system("streamlit run /content/modelUsingGpt-neo-2.7B.py")

# Run Streamlit in the background
thread = threading.Thread(target=run_streamlit)
thread.start()

# Wait a few seconds to ensure Streamlit has time to start
time.sleep(5)

# Create a public URL with an explicit protocol
url = ngrok.connect(8501, "http")
print(f"Streamlit is running at {url}")
```
